### 1. create the  dataframe from csv file

```python
from pyspark.sql import SparkSession
```

```python
spark=SparkSession.builder \
  .appName('MyApp') \
  .getOrCreate()
```

```python
df=spark.read \
  .format('csv') \
  .option('header', True) \
  .option('inferSchema', True) \
  .load('/content/sample_data/data.csv')
```

```python
df.show()
```

```python
df.printSchema()
```

### 2. Create dataframe from .csv file with deliliter "|"

### 3. Defining custom schema for csv file

```python
from pyspark.sql import SparkSession
```

```python
spark=SparkSession.builder \
  .appName('MyApp') \
  .getOrCreate()
```

```python
from pyspark.sql.types import *

schema=StructType([
    StructField("id", IntegerType(), True),
    StructField("name", StringType(), True),
    StructField("salary",IntegerType(), True)
])
```

```python
df=spark.read \
  .format('csv')  \
  .schema(schema) \
  .load('/content/sample_data/data_without_header.csv')
```

```python
df.show()
```
