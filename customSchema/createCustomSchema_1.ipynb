### Create dataframe from the list with Custom Schema

1. data = [(1,"Adi"),(2,"biba"),(3,"conde")]  





````python
from pyspark.sql import SparkSession
from pyspark.sql.types import *
````

````python
spark=SparkSession.builder \
  .appName('MyApp') \
  .getOrCreate()
````

````python
data = [(1,"Adi"),(2,"biba"),(3,"conde")]
````

````python
schema=StructType([
    StructField('id', IntegerType(), True),
    StructField('name', StringType(), True)
])
````

````python
df=spark.createDataFrame(data,schema=schema)
````

````python
df.show()
````

2. data=[("subhag","M",30),("priyanka","F",28)]  

````python
from pyspark.sql import SparkSession
from pyspark.sql.types import *
````

````python
from pyspark.sql import SparkSession
from pyspark.sql.types import *
````

````python
data = [
        ("subhag","kumar","gupta",123,'M',25000.12),
        ("govind","charan","gupta",234,'M',15000.23),
        ("krishna","govind","gupta",345,'F',27000.45),
        ("ram","shree","gupta",456,'F',37000.67)
        ]
````

````python
schema=StructType([
    StructField('first_name', StringType(), True),
    StructField('middle_name', StringType(), True),
    StructField('last_name', StringType(), True),
    StructField('id', IntegerType(), True),
    StructField('sex', StringType(), True),
    StructField('salary', DoubleType(), True)
])
````

````python
df=spark.createDataFrame(data,schema)
````

````python
from pyspark.sql import SparkSession
from pyspark.sql.types import *
````

````python
data = [
    (("James","grey","Smith"),36636,"M",3100),
    (("Michael","Rose","nancy"),40288,"M",4300),
    (("Robert","john","Williams"),42114,"M",1400),
    (("Maria","Anne","Jones"),39192,"F",5500),
    (("Jen","Mary","Brown"),1234,"F",-1)
  ]
````

````python
schema=StructType([
    StructField('name', StructType([
        StructField('first_name', StringType(), True),
        StructField('middle_name', StringType(), True),
        StructField('last_name', StringType(), True)
    ])),
    StructField('salary', IntegerType(), True),
    StructField('sex', StringType(), True),
    StructField('id', IntegerType(), True)
])
````

````python
df=spark.createDataFrame(data,schema)
````

````python
df.show(truncate=False)
````

```
5. data = [
        (1, {'name': 'Italy', 'capital': 'Rome'}, 'euro'),
        (2, {'name': 'France', 'capital': 'Paris'}, 'euro'),
        (3, {'name': 'Japan', 'capital': 'Tokyo'}, 'yen')
    ]
```



````python
from pyspark.sql import SparkSession
from pyspark.sql.types import *
````

````python
spark=SparkSession.builder \
  .appName('MyApp') \
  .getOrCreate()
````

````python
data = [
        (1, {'name': 'Italy', 'capital': 'Rome'}, 'euro'),
        (2, {'name': 'France', 'capital': 'Paris'}, 'euro'),
        (3, {'name': 'Japan', 'capital': 'Tokyo'}, 'yen')
    ]
````

````python
schema=StructType([
    StructField('id', IntegerType(), True),
    StructField('country', StructType([
        StructField('name', StringType(), True),
        StructField('capital', StringType(), True)
    ])),
    StructField('currency', StringType(), True)
])
````

````python
df=spark.createDataFrame(data,schema)
````

````python
df.show()
````
