### Create dataframe from the list with Custom Schema


```
7. data = [
  ("James",[["Java","Scala","C++"],["Spark","Java"]]),
  ("Michael",[["Spark","Java","C++"],["Spark","Java"]]),
  ("Robert",[["CSharp","VB"],["Spark","Python"]])
]
```



````python
from pyspark.sql import SparkSession
from pyspark.sql.types import *
````

````python
spark=SparkSession.builder \
  .appName('MyApp') \
  .getOrCreate()
````

````python
data = [
  ("James",[["Java","Scala","C++"],["Spark","Java"]]),
  ("Michael",[["Spark","Java","C++"],["Spark","Java"]]),
  ("Robert",[["CSharp","VB"],["Spark","Python"]])
]
````

````python
schema=StructType([
    StructField('name', StringType(), True),
    StructField('skills',ArrayType(
        ArrayType(StringType())), True)
])
````

````python
df=spark.createDataFrame(data,schema)
````

````python
df.show(truncate=False)
````

````python
spark.stop()
````

````python
from pyspark.sql import SparkSession
from pyspark.sql.types import *
````

````python
spark=SparkSession.builder \
  .appName('MyApp') \
  .getOrCreate()
````

````python
data = [
    (("James","","Smith"),["Java","Scala","C++"],"OH","M",1000),
    (("Anna","Rose",""),["Spark","Java","C++"],"NY","F",2000),
    (("Julia","","Williams"),["CSharp","VB"],"OH","F",3000),
    (("Maria","Anne","Jones"),["scirpt","VB"],"NY","M",4000),
    (("Jen","Mary","Brown"),["CSharp","VB"],"NY","M",5000),
    (("Mike","Mary","Williams"),["Python","VB"],"OH","M",6000)
 ]
````

````python
schema=StructType([
    StructField('full_name', StructType([
        StructField('fiest_name', StringType(), True),
        StructField('middle_name', StringType(), True),
        StructField('last_name', StringType(), True)
    ])),
    StructField('skills', ArrayType(StringType()), True),
    StructField('city', StringType(), True),
    StructField('sex', StringType(), True),
    StructField('salary', IntegerType(), True)
])
````

````python
df=spark.createDataFrame(data,schema)
````

````python
df.show(truncate=False)
````
