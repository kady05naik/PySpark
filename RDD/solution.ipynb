### 1. Create RDD from List

```python
from pyspark.sql import SparkSession
```

```python
spark=SparkSession \
  .builder \
  .appName('MyApp')  \
  .getOrCreate()
```

```python
lst=[1,9,5,6,8,9,0,3]
```

```python
rdd=spark.sparkContext.parallelize(lst)
```

```python
print(rdd.collect())
```

## Convert rdd to dataframe

```python
from pyspark.sql import SparkSession
```

```python
spark=SparkSession \
  .builder \
  .appName('MyApp') \
  .getOrCreate()
```

```python
from pyspark.sql.types import *
```

```python
schema=StructType([
    StructField("id", IntegerType(), True),
    StructField("name", StringType(), True),
    StructField("salary", IntegerType(), True)
])
```

```python
rdd_split=rdd.map(lambda x : x.split(','))
print(rdd_split.collect())
```

In above rdd_split [['1', 'John', '5000'], ['2', 'Alice', '6000'], ['3', 'Bob', '7000']] 1 and 5000 also coverted into string. So first we will convert it into int and then we will create an  dataframe


```python
rdd_typed=rdd_split.map(lambda x : (int(x[0]), x[1], int(x[2])))
print(rdd_typed.collect())
```

Now values converted to integer [(1, 'John', 5000), (2, 'Alice', 6000), (3, 'Bob', 7000)]

```python
df=spark.createDataFrame(rdd_typed,schema=schema)
df.show()
df.printSchema()
```
