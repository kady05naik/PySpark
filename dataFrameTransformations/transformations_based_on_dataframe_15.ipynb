````python
from pyspark.sql import SparkSession
````

````python
spark=SparkSession.builder.appName('MyApp').getOrCreate()
````

````python
sample1 = [(1,"Sagar","CSE","UP",89),
                (2,"Shivam","IT","MP",86),
                (3,"Muni","MTech","AP",70)
               ]
````

````python
sample2 = [(5,"Raj","CSE","HP"),
                (6,"Kunal","Mech","JK")
               ]
````

````python
from pyspark.sql.types import *
````

````python
schema1=StructType([
    StructField("ID", IntegerType(), True),
    StructField("Student_Name", StringType(), True),
    StructField("Dept_Name", StringType(), True),
    StructField("State", StringType(), True),
    StructField("Marks", IntegerType(),True)
])
````

````python
schema2=StructType([
    StructField("ID", IntegerType(), True),
    StructField("Student_Name", StringType(), True),
    StructField("Dept_Name", StringType(), True),
    StructField("State", StringType(), True),
])
````

````python
df1=spark.createDataFrame(sample1, schema1)
df2=spark.createDataFrame(sample2, schema2)
````

### Merge two Dataframes using PySpark (if both dataframe has different number of coulmns

```
+---+------------+---------+-----+-----+
| ID|Student_Name|Dept_Name|State|Marks|
+---+------------+---------+-----+-----+
|  1|       Sagar|      CSE|   UP|   89|
|  2|      Shivam|       IT|   MP|   86|
|  3|        Muni|    MTech|   AP|   70|
|  5|         Raj|      CSE|   HP| null|
|  6|       Kunal|     Mech|   JK| null|
+---+------------+---------+-----+-----+
```



````python
df1.unionByName(df2,allowMissingColumns=True).show()
````
