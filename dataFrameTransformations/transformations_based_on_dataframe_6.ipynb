# Transformations on Data Frame 6

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
spark=SparkSession.builder \
  .appName('MyApp') \
  .getOrCreate()
```

```python
simpleData = [
    ("James","Sales","NY",90000,34,10000), \
    ("Michael","Sales","NY",86000,56,20000), \
    ("Robert","Sales","CA",81000,30,23000), \
    ("Maria","Finance","CA",90000,24,23000), \
    ("Raman","Finance","CA",99000,40,24000), \
    ("Scott","Finance","NY",83000,36,19000), \
    ("Jen","Finance","NY",79000,53,15000), \
    ("Jeff","Marketing","CA",80000,25,18000), \
    ("Kumar","Marketing","NY",91000,50,21000) \
  ]
```

```python
schema=StructType([
      StructField('name', StringType(), True),
      StructField('dept', StringType(), True),
      StructField('state', StringType(), True),
      StructField('salary', IntegerType(), True),
      StructField('age', LongType(), True),
      StructField('bonus', LongType(), True)
])
```

```python
df=spark.createDataFrame(simpleData, schema)
```



---



1️⃣ Order by name ascending

```python
df.orderBy(col('name')).show()
```



---



2️⃣ Order by name descending

```python
df.orderBy(col('name').desc()).show()
```



---



3️⃣ Order by two columns (department, state)

```python
df.orderBy(col('dept'), col('state')).show()
```



---



4️⃣ Department ascending and state descending

```python
df.orderBy(col('dept'), col('state').desc()).show()
```



---



1️⃣ Total salary department wise

```python
df.groupBy(col('dept')).agg(sum(col('salary')).alias('total_salary')).show()
```



---



2️⃣ Maximum salary

```python
df.agg(max(col('salary'))).show()
```



---



3️⃣ Department wise count of state

```python
df.groupBy('dept','state') \
  .agg(count(col('state')).alias('count_of_states')) \
  .show()
```



---



4️⃣ Max salary department wise

```python
df.groupBy('dept') \
  .agg(max(col('salary')).alias('maximum_salary')) \
  .show()
```



---



5️⃣ Min salary department wise

```python
df.groupBy('dept') \
  .agg(min(col('salary')).alias('min_salary')) \
  .show()
```



---



6️⃣ Count department wise

```python
df.groupBy('dept') \
  .count() \
  .show()
```



---



7️⃣ Group by multiple columns (department,state)
Find total salary and total bonus

```python
df.groupBy('dept','state') \
  .agg(
      sum(col('salary')).alias('total_salary'),
      sum(col('bonus')).alias('bonus')
      ).show()
```



---



9️⃣ Department & state total salary > 95000

```python
df.groupBy('dept', 'state') \
  .agg(sum(col('salary')).alias('total_salary')) \
  .filter(col('total_salary')>95000) \
  .show()
```
