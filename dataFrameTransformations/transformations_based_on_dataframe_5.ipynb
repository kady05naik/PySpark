# Transformations on Data Frame 5

````python
from pyspark.sql import SparkSession
````

````python
spark=SparkSession\
  .builder\
  .appName('MyApp') \
  .getOrCreate()
````

````python
data=data = [("James", "Sales", 3000), \
    ("Robert", "Sales", 3000), \
    ("Jeff", "Marketing", 4000), \
    ("Maria", "Finance", 3000), \
    ("James", "Sales", 3000), \
    ("Scott", "Finance", 3900), \
    ("Jen", "Finance", 3900), \
    ("Jeff", "Marketing", 4000), \
    ("Kumar", "Marketing", 4000), \
    ("Scott", "Finance", 3000) \
  ]
````

````python
from pyspark.sql.types import *
````

````python
schema=StructType([
    StructField('name',StringType(), True),
    StructField('dept', StringType(), True),
    StructField('salary', IntegerType(), True)
])
````

````python
df=spark.createDataFrame(data, schema)
````



---



1️⃣ Difference between distinct() and dropDuplicates()

distinct() -removes duplicate rows considering all columns.

````python
df.distinct() \
  .show()
````

dropDuplicates()-removes duplicates based only on department and salary.

````python
df.dropDuplicates(['dept', 'salary']).show()
````



```
distinct() removes duplicate rows from the entire DataFrame,
while dropDuplicates() allows removing duplicates based on specific columns.

Which is better: distinct() or dropDuplicates()?
Use distinct() when removing duplicates across all columns
Use dropDuplicates() when removing duplicates based on specific columns
```





---



2️⃣ Get only the distinct records

````python
df.distinct()\
  .show()
````



---



3️⃣ Apply distinct on selected columns (department, salary)

````python
from pyspark.sql.functions import *
````

````python
df.select(col('dept'), col('salary')) \
  .distinct() \
  .show()
````



---



4️⃣ Apply dropDuplicates on department and salary

````python
df.dropDuplicates(['dept','salary']) \
  .show()
````

````python
spark.stop()
````



---



## **TakeAway:**


**select().distinct()** returns unique combinations of selected columns and removes other columns.



**dropDuplicates()** removes duplicate rows based on specific columns but keeps the full row structure.
