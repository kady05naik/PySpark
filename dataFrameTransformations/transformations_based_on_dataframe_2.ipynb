# Transformations on Data Frame 2

````python
from pyspark.sql import SparkSession
````

````python
spark=SparkSession.builder \
  .appName('myApp') \
  .getOrCreate()
````

````python
data = [(1, "James", "Sales"),(2, "Michael", "Finance"),(3, "James", "Sales"),
         (4, "Maria", "HR"), (5, "Michael", "Finance"), (6, "Tom", "Sales"),
         (7, "Sara", "HR"), (8, "David", "Finance"),(9, "Mike", "Sales"),
         (10, "Emma", "HR")  ]
````

````python
from pyspark.sql.types import *
````

````python
schema=StructType([
    StructField('id',IntegerType(), True),
    StructField('name', StringType(), True),
    StructField('dept', StringType(), True)
])
````

````python
df=spark.createDataFrame(data,schema)
````

````python
df.show()
````
