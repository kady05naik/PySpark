# Transformations on Data Frame 10

### **Self Join in pyspark**

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
spark=SparkSession.builder \
  .appName('Myapp') \
  .getOrCreate()
```

```python
data= [
        (1,'Joe',70000,3),
        (2,'Henry',80000,4),
		    (3,'Sam',60000,None),
        (4,'Max',90000,None)
        ]
```

```python
schema=StructType([
    StructField('id', IntegerType(), True),
    StructField('name', StringType(), True),
    StructField('salary', LongType(), True),
    StructField('managerid', IntegerType(), True)
])
```

# Problem

We need to **find employee name and manager name**.

The managerId refers to id of another employee, so we **join the table with itself**.

```python
df=spark.createDataFrame(data,schema)

emp=df.alias('emp')
mgr=df.alias('mgr')
```

```python
from pyarrow import MemoryManager
result=emp.join(mgr, col('emp.managerid')==col('mgr.id'), "left") \
   .select(col('emp.name').alias('employee_name'), col('mgr.name').alias('manager_name'))
result.show()
```

```python
spark.stop()
```

# **Why do we use alias in self join?**


### **Because both tables have the same schema, alias helps Spark distinguish between the two instances of the same DataFrame.**
