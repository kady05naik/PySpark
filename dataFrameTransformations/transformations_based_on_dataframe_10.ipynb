# Transformations on Data Frame 10

````python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *
````

````python
spark=SparkSession.builder \
  .appName('myApp') \
  .getOrCreate()
````

````python
data_1 = [
    (105, "Praveen", 6000, 10),
    (102, "Krishna", 7000, 20),
    (103, "Mohan", 6000, 20),
    (106, "Radhe", 9000, 40)
]
````

````python
data_2 = [
    (102, "Krishna", 4000, 20),
    (103, "Mohan", 6000, 10),
    (104, "Anand", 3000, 30),
    (101, "Ravi", 5000, 10)
]
````

````python
schema=StructType([
    StructField('id',IntegerType(), True),
    StructField('name', StringType(), True),
    StructField('salary',IntegerType(), True),
    StructField('managerid', IntegerType(), True)
])
````

````python
df_emp1=spark.createDataFrame(data_1,schema)
````

````python
df_emp2=spark.createDataFrame(data_2,schema)
````

### If both dataframe have the same column name then how would you perform the join and select it
```
+-----+-------+------+------+
|EmpID|  Ename|Salary|DeptNo|
+-----+-------+------+------+
|  105|Praveen|  6000|    10|
|  102|Krishna|  7000|    20|
|  103|  Mohan|  6000|    20|
|  106|  Radhe|  9000|    40|
+-----+-------+------+------+

+-----+-------+------+------+
|EmpID|  Ename|Salary|DeptNo|
+-----+-------+------+------+
|  102|Krishna|  4000|    20|
|  103|  Mohan|  6000|    10|
|  104|  Anand|  3000|    30|
|  101|   Ravi|  5000|    10|
+-----+-------+------+------+
```


````python
df_emp1.join(df_emp2,'id','inner') \
  .select(
          df_emp1.id.alias('emp1_id'),
          df_emp1.name.alias('emp1_name'),
          df_emp1.salary.alias('emp1_salary'),
          df_emp1.managerid.alias('emp1_managerid'),
          df_emp2.id.alias('emp2_id'),
          df_emp2.name.alias('emp2_name'),
          df_emp2.salary.alias('emp2_salary'),
          df_emp2.managerid.alias('emp2_managerid'),
          ) \
  .show()
````
