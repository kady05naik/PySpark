````python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *
````

## **Pivot**
In PySpark is used to transform rows into columns by grouping data and applying aggregation functions.

It is typically used with **groupBy()** and **pivot()** to create summarized tables.



```
syntax -
df.groupBy("group_column") \
  .pivot("pivot_column") \
  .agg(aggregation_function)
```


When Pivot is Used

1.   Creating summary tables
2.   Converting categorical values into columns
3.   Building reporting datasets
4.   Preparing data for BI tools (Power BI / Tableau)

````python
df.groupBy('dept') \
  .pivot('gender') \
  .count() \
  .show()
````

````python
data=[("Banana",1000,"USA"),("Banana",1000,"USA"), ("Carrots",1500,"USA"), ("Beans",1600,"USA"), \
      ("Orange",2000,"USA"),("Orange",2000,"USA"),("Banana",400,"China"), \
      ("Carrots",1200,"China"),("Beans",1500,"China"),("Orange",4000,"China"), \
      ("Banana",2000,"Canada"),("Carrots",2000,"Canada"),("Beans",2000,"Mexico")]
````

````python
schema=StructType([
    StructField('product', StringType(), True),
    StructField('amount', IntegerType(), True),
    StructField('country', StringType(), True)
])
````

````python
df=spark.createDataFrame(data, schema)
````

````python
df.groupBy('product') \
  .pivot('country') \
  .agg(sum(col('amount'))) \
  .show()
````

````python
data = [
    ("A", "North", 500, "2023-01-01"),
    ("B", "South", 300, "2023-01-02"),
    ("A", "South", 700, "2023-01-02"),
    ("B", "North", 400, "2023-01-03"),
    ("A", "East", 600, "2023-01-04"),
    ("B", "East", 250, "2023-01-05")
]
````

````python

schema = StructType([
    StructField("product", StringType(), True),
    StructField("region", StringType(), True),
    StructField("amount", IntegerType(), True),
    StructField("date", StringType(), True)
])
````

````python
df=spark.createDataFrame(data,schema)
````

**Do needful for below output**

```
+---------+-------+-------+-----+
| product | North | South | East|
+---------+-------+-------+-----+
| A       | 500   | 700   | 600 |
| B       | 400   | 300   | 250 |
+---------+-------+-------+-----+
```



````python
df.groupBy('product') \
  .pivot('region') \
  .agg(sum(col('amount'))) \
  .orderBy(col('product')) \
  .show()
````
