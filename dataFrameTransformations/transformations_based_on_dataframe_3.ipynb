```python
from pyspark.sql import SparkSession
from pyspark.sql.types import *
```

```python
spark=SparkSession.builder \
  .appName('MyApp') \
  .getOrCreate()
```

```python
data = [
    (1, 704, "STANDARD", None, "PR", 3010),
    (2, 704, None, "PASEO COSTA DEL SUR", "PR", None),
    (3, 709, None, "BDA SAN LUIS", "PR", 3700),
    (4, 76166, "UNIQUE", "CINGULAR WIRELESS", "TX", 840),
    (5, 76177, "STANDARD", None, "TX", None)
]
```

```python
schema=StructType([
    StructField('id',IntegerType(), True),
    StructField('pincode', IntegerType(), True),
    StructField('type',StringType(), True),
    StructField('city', StringType(), True),
    StructField('state', StringType(), True),
    StructField('population',IntegerType(), True)
])
```

```python
df=spark.createDataFrame(data, schema)
```



---



### filter those record that 'type' column value is null

```python
from pyspark.sql.functions import *
df.filter(col('type').isNull()).show()
```



---



### filter those record that type and population column value is not null

```python
df.filter((col('type').isNotNull()) & (col('population').isNotNull())).show()
```



---



### filter those record that type or population column value is not null

```python
df.filter(col('type').isNotNull() | col('population').isNotNull()).show()
```
