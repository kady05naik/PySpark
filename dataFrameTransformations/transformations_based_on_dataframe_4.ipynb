# Transformations on Data Frame 4

```python
from pyspark.sql import SparkSession
```

```python
spark=SparkSession.builder \
  .appName('myApp') \
  .getOrCreate()
```

```python
data = [
    (("James","","Smith"),["Java","Scala","C++"],"OH","M",1000),
    (("Anna","Rose",""),["Spark","Java","C++"],"NY","F",2000),
    (("Julia","","Williams"),["CSharp","VB"],"OH","F",3000),
    (("Maria","Anne","Jones"),["scirpt","VB"],"NY","M",4000),
    (("Jen","Mary","Brown"),["CSharp","VB"],"NY","M",5000),
    (("Mike","Mary","Williams"),["Python","VB"],"OH","M",6000)
 ]
```

```python
from pyspark.sql.types import *
```

```python
schema=StructType([
    StructField('name', StructType([
        StructField('fname', StringType(), True),
        StructField('mname', StringType(), True),
        StructField('lname', StringType(), True)
    ])),
    StructField('languages', ArrayType(StringType()), True),
    StructField('state', StringType(), True),
    StructField('gender', StringType(), True),
    StructField('salary', IntegerType(), True)
])
```

```python
df=spark.createDataFrame(data,schema)
```

1️⃣ Retrieve name, languages, state, salary

```python
df.select('name', 'languages', 'state' ) \
  .show(truncate=False)
```



---



2️⃣ Retrieve first name and last name from name

```python
df.select('name.fname', 'name.lname') \
  .show(truncate=False)
```



---



3️⃣ Retrieve first and second element of languages

```python
from pyspark.sql.functions import *
```

```python
df.select(
    col('languages')[0].alias('language_1'),
    col('languages')[1].alias('language_2')
    ).show()
```



---



4️⃣ Get records where state = OH

```python
df.filter(col('state')=='OH') \
  .show(truncate=False)
```



---



5️⃣ Get records where state is NOT OH

```python
df.filter(col('state')!='OH') \
  .show()
```



---



6️⃣ Get records where salary > 5000

```python
df.filter(col('salary')>5000) \
  .show(truncate=False)
```



---



7️⃣ Get records where state = NY AND gender = M

```python
df.filter((col('state')=='NY') & (col('gender')=='M')) \
  .show()
```



---



8️⃣ Get records where salary > 2000 OR gender = M

```python
df.filter((col('salary')>2000) | (col('gender')=='M')) \
  .show()
```



---



9️⃣ Get records where salary between 2000 and 5000

```python
df.filter(col('salary').between(2000,5000)) \
  .show()
```



---



🔟 Get records where salary > 5000 AND state = OH

```python
df.filter((col('salary')>5000) & (col('state')=='OH')) \
  .show()
```



---



1️⃣1️⃣ Get records where salary > 1000 AND first name starts with J

```python
df.filter((col('salary')>1000) & (col('name.fname').startswith('J'))) \
          .show(truncate=False)
```
