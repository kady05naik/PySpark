# Transformations on Data Frame 9

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
spark=SparkSession.builder \
  .appName('MyApp') \
  .getOrCreate()
```

```python
emp_data = [
    (1, "James", 101),
    (2, "Michael", 102),
    (3, "Robert", 103),
    (4, "Maria", None ),
    (5, "Jen", None)
]
```

```python
dept_data = [
    (101, "Sales"),
    (102, "Finance"),
    (103, "Marketing"),
    (108, "Marketing")
]
```

```python
emp_schema=StructType([
    StructField('id',IntegerType(), True),
    StructField('name', StringType(), True),
    StructField('dept_id', IntegerType(), True)
])
```

```python
dept_schema=StructType([
    StructField('dept_id',IntegerType(), True),
    StructField('dept_name',StringType(), True)
])
```

```python
emp_df=spark.createDataFrame(emp_data,emp_schema)
```

```python
dept_df=spark.createDataFrame(dept_data, dept_schema)
```



---



# INNER JOIN

Returns only matching records from both tables

```python
emp_df.join(dept_df,'dept_id','inner').show()
```



---



# LEFT OUTER JOIN

Returns all records from the left table and matching records from the right table.

```python
emp_df.join(dept_df,'dept_id','left').show()
```



---



# RIGHT OUTER JOIN

Returns all records from the right table and matching records from the left.

```python
emp_df.join(dept_df, 'dept_id','right').show()
```



---



# LEFT SEMI JOIN

Definition:


leftsemi returns **only** the **matching records** from the **left** table.

```python
emp_df.join(dept_df,'dept_id','leftsemi').show()
```



---



# LEFT ANTI JOIN


Definition

leftanti returns **only non-matching records** from the **left** table.

Meaning: rows in left table not present in right table.

```python
emp_df.join(dept_df,'dept_id','leftanti').show()
```

```python
spark.stop()
```
