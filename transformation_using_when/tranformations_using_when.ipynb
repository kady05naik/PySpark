# Transformations using **when**

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
spark=SparkSession.builder \
  .appName('MyApp') \
  .getOrCreate()
```

```python
data = [("Saif", "", "Shaikh", "36636", "M", 60000),
            ("Ram", "Shirali", "", "40288", "M", 70000),
            ("Aniket", "", "Mishra", "42114", "", 400000),
            ("Mitali", "Sahil", "Kashiv", "39192", "F", 500000),
            ("Nahid", "Alim", "Shaikh", "", "F", 0)]
```

```python
schema=StructType([
    StructField('fname', StringType(), True),
    StructField('mname',StringType(),True),
    StructField('lname', StringType(), True),
    StructField('nos', StringType(), True),
    StructField('gender', StringType(), True),
    StructField('salalry', IntegerType(), True)
])
```

```python
df=spark.createDataFrame(data,schema)
```

```python
df.show()
```

**1- update the value of gender column when gender is M then Male , if geneder is F then Female else unknown**





```python
df.withColumn('gender',when(col('gender')=='M',lit('Male')) \
              .when(col('gender')=='F', lit('Female')) \
              .otherwise(lit('Unknown'))) \
              .show()
```

**2- update the value of gender column when gender is M then Male , if geneder id F then Female else blank**

```python
df.withColumn('gender', when(col('gender')=="M",lit('Male')) \
              .when(col('gender')=='F', 'Female') \
              .otherwise("")) \
              .show()
```

**Create a new column gender and populate the value from sex column in such a way if sex column value is M then Male ,if F then Female else blank and at the last drop the sex column**

```python
data = [("James","M"),("Michael","F"),("Jen","")]
```

```python
schema=StructType([
    StructField('name', StringType(), True),
    StructField('sex', StringType(),True)
])
```

```python
df=spark.createDataFrame(data, schema)
```

```python
df.show()
```

```python
df.withColumn('gender',when(col('sex')=='M', lit('Male')) \
              .when(col('sex')=='F', lit('Female')) \
              .otherwise('')) \
              .drop(col('sex')) \
              .show()
```

**Create a new column discount based on the salary  named is discount, if salary is Salary <= 50000 AND Salary >= 40000 then add  10 else 20**

```python
data = [("111",50000),("222",60000),("333",40000)]
```

```python
schema=StructType([
    StructField('empid', StringType(), True),
    StructField('sal', IntegerType(), True)
])
```

```python
df=spark.createDataFrame(data, schema)
```

```python
df.withColumn('discount',when((col('sal')>=40000) & (col('sal')<=50000), col('sal')+10) \
              .otherwise(col('sal')+20))\
              .show()
```

```python
salary_data = [
    (1, "James", 30000),
    (2, "Michael", 60000),
    (3, "Robert", 45000),
    (4, "Maria", 120000),
    (5, "Jen", 20000)
]
```

```python
schema=StructType([
    StructField('id', IntegerType(), True),
    StructField('name', StringType(), True),
    StructField('salary', IntegerType(), True)
])
```

```python
df=spark.createDataFrame(salary_data,schema)
```

```python
df.show()
```

**If the salary is greater than 50,000, classify as High.If the salary is between 30,000 and 50,000, classify as Medium. Otherwise, classify as Low.**

```python
df.withColumn('classification', when(col('salary')>50000, lit('High')) \
              .when(col('salary').between(30000,50000), lit('Medium')) \
              .otherwise(lit('Low'))) \
              .show()
```

```python
data = [
    (1, 60000, "East"),
    (2, 40000, "West"),
    (3, 15000, "East"),
    (4, 30000, "West"),
    (5, 70000, "East"),
    (6, 25000, "West")
]
```

```python
schema=StructType([
    StructField('id', IntegerType(), True),
    StructField('sales_amount', IntegerType(), True),
    StructField('region', StringType(), True)
])
```

```python
df=spark.createDataFrame(data, schema)
```

```python
df.show()
```

**If the sales_amount is greater than 50,000 and the region is East, classify as Top Performer. If the sales_amount is between 20,000 and 50,000 and the region is West, classify as Average Performer. Otherwise, classify as Under Performer.**

```python
df.withColumn('classification', when((col('sales_amount')>50000) & (col('region')=='East'), lit('Top Performer')) \
              .when((col('sales_amount').between(20000,50000)) & (col('region')=='West'), lit('Average Performance')) \
              .otherwise(lit('Under Performer'))) \
              .show()
```

```python
data = [
    (1, "Alice", 95),
    (2, "Bob", 85),
    (3, "Charlie", 65),
    (4, "David", 45),
    (5, "Eve", None)
]
```

```python
schema=StructType([
    StructField('id', IntegerType(), True),
    StructField('name', StringType(), True),
    StructField('marks', IntegerType(), True)
])
```

```python
df=spark.createDataFrame(data,schema)
```

```python
df.show()
```

**If the marks are greater than 90, classify as Excellent. If the marks are between 70 and 90, classify as Good. If the marks are between 50 and 70, classify as Average. If the marks are less than 50, classify as Needs Improvement.**

```python
df.withColumn('remark',when(col('marks')>90, lit('Excellent')) \
              .when(col('marks').between(70,90),lit('Good')) \
              .when(col('marks').between(50,70),lit('Average')) \
              .when(col('marks')<50,lit('Needs Improvement'))) \
              .show()
```
